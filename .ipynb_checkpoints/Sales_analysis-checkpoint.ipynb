## Importing libraries

```python
import pandas as pd
import glob
```

### Creating a DataFrame with 12 csv Files

```python
path = 'Sales_Data'
csv_files = glob.glob(path + '/*.csv')

df_list = (pd.read_csv(file) for file in csv_files)

df = pd.concat(df_list, ignore_index=True)
df
```

### Clean Data (Drop NAN rows)

```python
df.dropna(inplace=True)
```

#### Drop Duplicated column headers

```python
df = df[df['Order Date'].str[0:2] != 'Or']
```

#### Convert colunms with string numbers to correct type

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

### Augmenting data with more columns

#### Adding a month column (To see the best selling month)

```python
df['month'] = df['Order Date'].str[0:2]
df
```

```python
df['month'] = df['month'].astype('int32')
df.head()
```

```python
df['sales'] = df['Quantity Ordered'] * df['Price Each']
df.head()
```

#### Adding a city colunm

```python
df['City'] = df['Purchase Address'].apply(lambda x: x.split(',')[1])
df['State'] = df['Purchase Address'].apply(lambda x: x.split(',')[2].split(' ')[1])
```

```python
df.head()
```

### What was the best selling month? How much was it?

```python
results = df.groupby('month').sum()
results
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1, 13)

plt.bar(months, results['sales'])
plt.xticks(months)
plt.ylabel('Sales in USD $')
plt.xlabel('Month')
plt.ticklabel_format(useOffset=False, axis='y', style='plain')
plt.show()
```

### What City had the highest number in sales

```python
results_cities = df.groupby('City').sum()
results_cities
```

```python
import matplotlib.pyplot as plt

cities = [city for city, df in df.groupby('City')]

plt.bar(cities, results_cities['sales'])
plt.xticks(cities, rotation='vertical')
plt.ylabel('Sales in USD $')
plt.xlabel('City name')
plt.ticklabel_format(useOffset=False, axis='y', style='plain')
plt.show()
```

### What time should we display advertisements to maximize the likelihood of purchases?

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.head()
```

```python
df['Hour'] = df['Order Date'].dt.hour
df['Minute'] = df['Order Date'].dt.minute
df['count'] = 1
```

```python
hours = [hour for hour, df in df.groupby('Hour')]

plt.plot(hours, df.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()
```

### What products are most often sold together?

```python
new_df = df[df['Order ID'].duplicated(keep=False)]
new_df['Grouped'] = new_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
new_df = new_df[['Order ID', 'Grouped']].drop_duplicates()
new_df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in new_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
    print(key, value)
```
